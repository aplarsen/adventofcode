```python
# Read our input
with open ('input1.txt') as f:
    input = f.read().split('\n')
```

```python
# Set starting position
position = 50

# Initialize zero count
zeroes = 0

# Loop instruction
for instruction in input:
    # Split the instruction into a direction and distance
    direction = instruction[0]
    distance = int(instruction[1:])

    # Loop the clicks
    for i in range(distance):
        # Increment or decrement based on direction
        if direction == 'L':
            position = position - 1
        if direction == 'R':
            position = position + 1

        # Reset based on bounds
        if position < 0:
            position = 99
        if position > 99:
            position = 0
    
    # Increment zero count if we landed on 0
    if position == 0:
        zeroes = zeroes + 1

zeroes
```
