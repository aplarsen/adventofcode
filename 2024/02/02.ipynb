```python
with open( 'input.txt' ) as f:
    input = f.readlines()

    input = [ list(map(int, row.strip().split(' '))) for row in input ]

input
```

```python
safes = 0

for row in input:
    safe = True

    direction = (row[1]-row[0] > 0)

    for l, level in enumerate(row[:-1]):
        if (row[l+1]-row[l]>0) != direction:
            safe = False
        if ( abs(row[l+1]-row[l]) not in [1,2,3] ):
            safe = False

    if safe:
        safes += 1
        
safes
```

```python
def row_safe(row):
    safe = True

    direction = (row[1]-row[0] > 0)

    for l, level in enumerate(row[:-1]):
        if (row[l+1]-row[l]>0) != direction:
            safe = False
        if ( abs(row[l+1]-row[l]) not in [1,2,3] ):
            safe = False
    
    return safe

safes = 0

for row in input:
    if row_safe(row):
        safes += 1

    else:
        for i in range(len(row)):
            if row_safe( row[0:i] + row[i+1:] ):
                safes +=1

                break


safes
```
