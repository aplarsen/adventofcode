```python
import re
from functools import reduce
```

```python
with open( 'input.txt' ) as f:
    input = f.read()
```

```python
matches = re.findall( r'mul\((\d+?),(\d+?)\)', input )

matches
```

```python
reduce(lambda x, y: x + int(y[0])*int(y[1]), matches, 0)
```

```python
# Part 2
matches = re.findall( r'mul\(\d+?,\d+?\)|do\(\)|don\'t\(\)', input )

matches
```

```python
total = 0

do = True

for match in matches:
    if match == 'do()':
        do = True
    elif match == 'don\'t()':
        do = False
    else:
        if do:
            if ( mult := re.match( r'mul\((\d+?),(\d+?)\)', match ) ):
                total += int(mult.groups()[0])*int(mult.groups()[1])

total
```
