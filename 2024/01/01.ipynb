```python
with open('input.txt') as f:
    input = f.read()

    input = [ tuple(map(int, line.split( '   '))) for line in input.split('\n') ]

input
```

```python
list1 = [ row[0] for row in input ]
list2 = [ row[1] for row in input ]

list1, list2
```

```python
list1.sort()

list2.sort()

list1, list2
```

```python
sum = 0

for i in range(len(list1)):
    sum += abs(list1[i] - list2[i])

sum
```

```python
similarity = 0

for i in range(len(list1)):
    similarity += list1[i] * len(list(filter(lambda x: x == list1[i], list2)))

similarity
```
